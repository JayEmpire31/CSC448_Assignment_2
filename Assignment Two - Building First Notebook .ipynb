## Testing out the pandas library using data from the MTA Subways Analytics

This line of code prints out the first five data entry points of the CSV file. This can be helpful to properly read the data.



```python
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv('MTA_recent_ridership_data_20210315.csv')
df.plot(kind ='scatter', x = 'Date', y = 'Subways: Total Estimated Ridership')
plt.xlim(7,1)
plt.show
```

The above code displays the ridership in the subways for one week of 3/8/2021 using the plt.xlim(7,1) code. A scatter plot was used to show the data.
